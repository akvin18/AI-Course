# Quiz: Pretraining and Self-Supervised Learning

### Multiple Choice Questions

1. What is the purpose of pretraining in modern NLP models?  
a) To speed up inference  
b) To eliminate the need for tokenization  
c) To learn general language representations from large corpora  
d) To compress the model size

2. What does BERT use as a pretraining objective?  
a) Next sentence prediction  
b) Masked language modeling  
c) Both A and B  
d) Translation prediction

3. What makes pretraining **self-supervised**?  
a) The model supervises itself during fine-tuning  
b) It uses labeled datasets only  
c) The training objective is generated from the data itself  
d) It doesn’t require tokenization

4. Which of the following is a **benefit** of pretraining a model before fine-tuning?  
a) It requires less memory  
b) It performs well out-of-the-box on many tasks  
c) It prevents overfitting  
d) It makes the model deterministic

5. What is the main difference between fine-tuning and pretraining?  
a) Pretraining uses task-specific data  
b) Fine-tuning is done before pretraining  
c) Pretraining is done on general data; fine-tuning adapts it to a specific task  
d) There is no difference


### Analytical Questions

2. What challenges can arise when transferring a pretrained model to a domain-specific task?

3. Explain the concept of transfer learning in NLP using an example involving pretraining and fine-tuning.

4. What’s the difference between autoregressive and masked language modeling objectives? When would you use each?

5. If a model is pretrained on social media posts and then fine-tuned on medical data, what are the risks and how can you mitigate them?
