## წრფივი რეგრესია

პირველ რიგში განვიხილოთ, უმარტივესი წრფივი რეგრესიის მოდელი. წრფივი რეგრესია არის მონაცემებზე წრფის მორგება:

საილუსტრაციოთ განვიხილოთ რეგრესიული ამოცანა, სადაც გვინდა ადამიანის გამოცდილებით სამსახურსი ვიპროგნოზოთ მისი ხელფასი.

პირველ რიგში გავავლოთ წრფე მონაცემებზე შემთხვევითად:

შემდგომ გავზომოთ მანძილი გავლებული წრფიდან მონაცემებამდე, ავიყვანოთ ეს მანძილები კვადრატში და ავჯამოთ:

შემდგომ შემოვატრიალოთ წრფე ცოტათი, კვლავ გავზომოთ მანძილები და დავთვალოთ მათი კვადრატების ჯამი:

გავიმეოროთ ეს პროცედურა მანამ სანამ არ მოვახდენთ წერტილების წრფესთან მანძილების კვადრატების ჯამის მინიმიზაციას:

მონაცემებზე წრიფს ასეთ მორგებას უწოდებენ "უმცირეს კვადრატთა" მეთოდს.

ჩვენ მივიღეთ მონაცემებზე მორგებული წრფე, რომელიც არის ჩვენი საპროგნოზო მოდელი:

პროგნოზირებისთვის ჩვენს მიერ აგებული წრფის განტოლებით შევძლებთ ადამიანის ხელფასის პროგნოზირებას. (X-ის ჩავსვამთ გამოცდილებას და  Y= 0.5 +0.8*X ის გამოყენებით მივიღებთ Y -ის მნიშვნელობას).

მაგრამ ბუნებრივია დავსვათ კითხვა თუ რამდენად კარგია ჩვენი პროგნოზი. ამისთვის ჩვენ უნდა დავთვალოთ $R^2$. მოდით ჩვენი მონაცემების მაგალითზე ვნახოთ თუ როგორ ითვლება $R^2$ და რას ნიშნავს ის.

პირველ რიგში დავთვალოთ ჩვენს მონაცემებში საშუალო ხელფასი. შემდგომ დავთვალოთ მანძილი ხელფასის თითოეულ მნიშვნელობასა და მათ საშუალოს შორის, ავიყვანოთ ეს მანძილები კვადრატში და ავჯამოთ. 

მოდით ეს ჯამი აღვნიშნოთ s(mean) -- "კვადრატების ჯამი საშუალოს გარშემო"

s(mean) = (მონაცემებს – საშუალო)$^2$

ვარიაცია საშუალოს გარშემო = $\frac{s(mean)}{n}$

$Var(mean) = \frac{s(mean)}{n}$

სადაც n არის მონაცემების (წერტილების) რაოდენობა. 

შემდგომ დავთვალოთ წრფის და წერტილების მანძილების კვადრატების ჯამი და ის აღვნიშნოთ s(line):

s(line) = (მონაცემებს – წრფე)$^2$.

როგორც საშუალოს შემთხვევაში ვარიანსი წრფის გარშემო იქნება:

$Var(line) = \frac{s(line)}{n}$

შესაბამისად ვღებულობთ $R^2$-ის ფორმულას:

$R^2 = \frac{Var(mean) - Var(line)}{Var(mean)}$

$R^2$ გვეუბნება ხელფასის ვარიაციის რა ნაწილი აიხსნება თუ გავითვალისწინებთ მუსაობის გამოცდილებას. 

დავუშვათ Var(mean)=10 , ხოლო  Var(line)=3, მაშინ $R^2$=0.7=70%

ეს ნიშნავს რომ სამუშაო გამოცდილება "ხსნის" 70% ხელფასის ვარიაციას.

სხვანაირად:

70% ხელფასის კვადრატების ჯამის შეიძლება აიხსნას სამუშაო გამოცდილებით.

$R^2$ შეგვიძლია გამოვიყენოთ რეგრესიული მოდელის წარმადობის საზომად.

წრფივი რეგრესიის გამოყენებისას ჩვენ შეგვიძლია ვისარგებლოთ არსებული ბიბლიოთეკებით, რომელსაც შეასრულებს ყველა გამოთვლებს.

```python
def replace_most_pop_miss(X):

    if len(list(X.select_dtypes(include=['object']).columns)) > 0:
        X.replace(dict.fromkeys(['?', 'n/a', 'nan', 'NaN', ',', 'N/a', 'N/A', 'none', '*', '.'], np.nan), inplace=True)


        X = X.replace(r'^\s*$', np.nan, regex=True)

    return X
```

```python
def check_columns_type_with_Nan_replace(X, column):
            # need to define int column with NaN(should be float type)
            X[column] = X[column].fillna(0)
            count_int = 0
            lenth = len(X[column])


            for i in X[column]:
                try:
                    if i%1 == 0:
                        count_int +=1
                except:
                    continue
            if count_int == lenth:
                
                X[column]=X[column].astype(int)
            return X[column]
```

```python
#დავწეროთ ფუნქცია რომელიც განგვიცალკევებს კადეგორიულ და რიცხვით ცვლადების შემდეგნაირად:
# თუ 40 ზე ნაკლები უნიკალური მნიშვნელობაა მაშინ ასეთი ცვლადი იყოს კატეგორიული
def define_columns_type(X):
    high_card_features_ls = []
    cat_features_ls = []
    num_features_ls = []
    
    X = pd.DataFrame.copy(X)
    for column in X.columns:     
            # if number of unique values more than 63, that column should be categorical
            # if column numerical
            if (X[column].dtype in [int, float]):
                
                
                # try to replace Nan and check column type
                X[column] = check_columns_type_with_Nan_replace(X, column)
                #check if column 'int'
                if X[column].dtype in [int]:
                    if X[column].nunique()>40:
                        num_features_ls.append(column)
                    else:
                        cat_features_ls.append(column)

                else:
                    num_features_ls.append(column)

            # check object columns   
            elif (X[column].nunique()>40):
                num_features_ls.append(column)
            else:
                cat_features_ls.append(column)

    return cat_features_ls, num_features_ls
```

## გრადიენტული დაბუსტვა (gradient boosting)

მოდით შევეხოთ ყველაზე გამოყენებად მოდელირების ალგორითმს "გრადიენტული დაბუსტვას" რეგრესიის შემთხვევაში (ანალოგიურად შეგვიძლია ის გამოვიყენოთ კლასიფიკაციის ამოცანებისთვის, როგორც უკვე განხილული ალგორითმები), შემდგომ კი მისი გამოყენებით ავაგოთ მოდელი ზემოთ შემოტანილ მონაცემებზე.

ჩვენ უკვე გვესმის გადაწყვეტილების ხეები. ეხლა ავხსნათ gradient boosting შემდეგ მონაცემებზე, სადაც უნდა ვიპროგნოზოთ ადამიანის წონა:

პირველ რიგში ვითვლით საშუალო წონას:

საშუალო წონა = 71.2

ეს არის პირველი მცდელობა ყველა ადამიანის წონის პროგნოზირების.

შემდეგი ნაბიჯია, რომ ავაგოთ გადაწყვეტილების ხე საშუალო წონის ცდომილებასთან მიმართებაში, ე.ი.

ამ მონაცემებზე გადაწყვეტილების ხე გამოიყურება შემდეგნაირად :

ცხადია რეგრეიის შემთხვევაში, ფოთლების მნიშვნელობებს სადაც რამოდენიმე ჩანაწერი მოხვდა საშუალოთი ვცვლით:

ჩვენ დავიწყეთ საწყისი პროგნოზით "საშუალო წონა = 71.2". იმისთვის რომ ავირიდოთ მოდელის ზედმეტად მორგებულობა უნდა შემოვიტანოთ learnig rate და განვაახლოთ პროგნოზირებული "წონა" (learning rate ავიღოთ 0.1):

პროგნოზირებული წონა = საშუალო წონა + learnig rate * ზედა ხის პასუხი


პროგნოზირებული წონა პირველი ჩანაწერისთვის = 71.2+0.1* 16.8= 72.9

როგორც ვხედავთ პროგნოზი ცოტათი უკეთესია ვიდრე საშუალოთი ნაპროგნოზები.

```python
import optuna
import pandas as pd
from sklearn import ensemble
from sklearn import datasets
from sklearn import model_selection
```

```python
#ალგორითმის ჰიპერპარამეტრების ოპტიმიზაცია
def objective(trial):
    # XGBoost parameters
    params = {
        "verbosity": 0,  
        "n_estimators": 1000,
        "max_depth": trial.suggest_int("max_depth", 4, 12),
        "learning_rate": trial.suggest_loguniform("learning_rate", 0.005, 0.09),
        "colsample_bytree": trial.suggest_loguniform("colsample_bytree", 0.2, 0.6),
        "subsample": trial.suggest_loguniform("subsample", 0.4, 0.8),
        "alpha": trial.suggest_loguniform("alpha", 0.01, 10.0),
        "lambda": trial.suggest_loguniform("lambda", 1e-8, 10.0),
        "gamma": trial.suggest_loguniform("lambda", 1e-8, 10.0),
        "min_child_weight": trial.suggest_loguniform("min_child_weight", 10, 1000)
    }

    model = XGBRegressor(**params)
    score = model_selection.cross_val_score(model, x_train, y_train, n_jobs=-1, cv=5, scoring="r2")
    r2_mean = score.mean()
    return r2_mean
```

```python
pruner = optuna.pruners.HyperbandPruner()
study = optuna.create_study(direction="maximize", pruner=pruner)
study.optimize(objective, n_trials=30, timeout=1800)
```

```python
study.best_params
```

```python
optimised_xgb = XGBRegressor(max_depth = study.best_params['max_depth'], learning_rate = 0.3,
                            colsample_bytree = study.best_params['colsample_bytree'], subsample = study.best_params['subsample'],
                            alpha = study.best_params['alpha'], Lambda = study.best_params['lambda'],
                            min_child_weight = study.best_params['min_child_weight'])

optimised_xgb.fit(x_train ,y_train)
```
