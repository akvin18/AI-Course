## გადაწყვეტილების ხეები (Decission trees)

გადაწყვეტილების ხეების ალგორითმი და მუშაობის პრინციპი ავხსნათ მაგალითის გამოყენებით.

განვიხილოთ შემდეგი მონაცემები, სადაც გვაქვს 3 ცვლადი და უნდა ვიპროგნოზოთ დადის თუ არა ადამიანი ცეკვაზე;

ჩვენ გამოვიყენებთ ამ მონაცემებს რომ ავაგოთ ეს კლასიფიკაციის ხე:

რომელიც პროგნოზირებს დადის თუ არა ვინმე ცეკვაზე.

ჩავთვალოთ, რომ ეს ხე არც გვინახავს და მხოლოდ მონაცემებით ავაგოთ ეს ხე.

პირველი რაც ჩვენ უნდა განვსაზღვროთ თუ რომელი მათგანი: "უყვარს მუსიკა", "უყვარს ცეკვა", "ასაკი", უნდა იყოს პირველი კითხვა რასაც ხის დასაწყისში დავსვამთ:

ეს რომ განვსაზღვროთ უნდა შევამოწმოთ "უყვარს მუსიკა" (და შემდგომ დანარჩენი ცვლადები) რამდენად კარგად პროგნოზირებს "დადის ცეკვაზეს".

ამისთვის უნდა ავაგოთ შემდეგი უმარტივესი ხეები:

მაგრამ როგორ გავზომოთ თუ რომელი ხე უკეთესად პროგნოზირებს საპროგნოზო ცვლადს ("დადის ცეკვაზე")?

რიცხვითად ეს იზომება Gini impurity–ით, რომელიც ითვლება შემდეგნაირად.

დავთვალოთ Gini impurity "უყვარს მუსიკისთვის". ის უნდა დავთვალოთ თითოეული ფოთოლისთვის (მწვანე უჯრებს ხეების დიაგრამაზე ამიერიდან ვუწოდებთ ფოთლებს).

Gini impurity პირველ ფოთოლში = 1 - ("კი"–ს ალბათობა)$^2$ – ("არა"–ს ალბათობა)$^2$ = $ 1 –\big(\frac{1}{1+3}\big)^2 - \big(\frac{3}{1+3}\big)^2= 0.375$

Gini impurity მეორე ფოთოლში = $ 1 –\big(\frac{2}{2+1}\big)^2 - \big(\frac{1}{2+1}\big)^2= 0.444$

ვინაიდან მარცხენა ფოთოლში არის 4 ადმაიანი ხოლო მარჯვენა ფოთოლში არის 3 ადამიანი, ფოთლები არ წარმოადგენენ ერთი და იგივე რაოდენობის ადამიანებს ამიტომაც სრული Gini impurity არის თითოეული ფოთოლის Gini impurity–ის  შეწონილი საშუალო. 

ხოლო წონებს თითოეული ფოთლისთვის ვთვლით შმედეგნაირად: ადამიანების რაოდენობა ფოთოლში შეფარდებული ადამიანების რაოდენობასთან ყველა ფოთოლში, შესაბამისად

წონა პირველი ფოთლის = $\frac{4}{4+3} $

წონა მეორე ფოთლის = $\frac{3}{4+3} $

აქედან გამომდინარე

სრული Gini impurity = წონა პირველი ფოთლის *  Gini impurity პირველ ფოთოლში +წონა მეორე ფოთლის * Gini impurity მეორე ფოთოლში = $\frac{4}{4+3} * 0.375 + \frac{3}{4+3}* 0.444 = 0.405 $

ე.ი.
Gini impurity "უყვარს მუსიკისთვის" = 0.405


ანალოგიურად მივიღებთ, რომ

Gini impurity "უყვარს ცეკვისთვის" = 0.214  (გამოთვალეთ და შეამოწმეთ).

Gini impurity "ასაკისთვის" ცოტათი განსხვავებულად ითვლება ვინაიდან ის არ არის ბინარული ცვლადი: ასაკის ცვლადს ვალაგებთ კლებადობით, თითოეული მეზობლებისთვის ვთვლით საშუალო მნიშვნელობებს, შემდგომ ვთვლით Gini impurity–ის თითოეული საშუალო ასაკისთვის საიდანაც უმცირესი მნიშვნელობა იქნება Gini impurity "ასაკისთვის".

ჩვენ უკვე შეგვიძლია გავცეთ პასუხი თუ რომელი კითხვა უნდა დავსვათ ხის დასაწყისში. ხეს დავიწყებთ იმ ცვლადით რომელსაც აქვს მინიმალური Gini impurity, ჩვენს შემთხევაში ეს არის "უყვარს ცეკვა":

4–ვე ადამიანი ვისაც უყვარს ცეკვა ამ node-ში არიან, 3 მათგანი დადის ცეკვაზე ხოლო 1 არა, რის გამოც ამ node–ს impurity  გააჩნია. ვეცადოთ შევამციროთ impurity, ცეკვის მოყვარულების გახლეჩვით ასაკის ან მუსიკის სიყვარულის მიხედვით:

ამისთვის ვკითხოთ იმ 4 ადამიანს ვისაც უყვარს ცეკვა უყვარს თუ არა მუსიკა:

ხოლო ამ გახლეჩვის სრული Gini impurity = 0.25. ხოლო ჩვენი მიზანია რომ ეს შევამციროთ.

ასაკიტ გახლეჩვისტვის ვკითხოთ იმ 4 ადამიანში, ასაკი= 12.5 გვაძლევს მინიმალურ impurity–ს, და შესაბამისად გავხლიცოთ 12.5–ით:

ანუ მივიღეთ რომ ასაკით გახლეჩვისას Gini impurity = 0.

შედეგად ჩვენი კლასიფიკაციის ხე იქნება:

ამით ჩვენ დავასრულეთ გადაწყვეტილების ხის აგება ჩვენს მონაცემებზე.

თუ ახალი ადამიანის მონაცემი გვაქვს და გვინდა ვიპროგნოზოთ დადის თუ არა ცეკვაზე:

ამ მონაცემებს გავატარებთ ჩვენს მიერ აგებულ ხეში. რადგანაც უყვარს ცეკვა გავყვებით ხეს მარცხნივ, და შემდგომ node-ზე რადგანაც ასაკი 15> 12.5 -ზე ხის დიაგრამა გვეტყვის პასუხს რომ დადის ცეკვაზე.

ჩვენ საფუძვლიანად ავხსენით გადაწყვეტილების ხის კლასიფიკაიის მოდელი. მაგრამ მისი გამოყენებისას ჩვენ შეგვიძლია ვისარგებლოთ არსებული ბიბლიოთეკებით, რომელსაც მივანდობთ ყველა გამოთვლებს.

```python
def replace_most_pop_miss(X):
    dict_nans = dict.fromkeys(['?', 'n/a', 'nan', 'NaN', ',', 'N/a', 'N/A', 'none', '*', '.'], np.nan)
    objects_list = X.select_dtypes(include = ['object']).columns
    X[objects_list] = X[objects_list].replace(dict_nans)
    return X
```

```python
df = replace_most_pop_miss(df)
```

მონაცემების დამუშავების დავში გამოყენებული ფუნქციით გადავყაროთ 70% ზე მეტი სიცარიელიანი სვეტები

```python
def more_70_of_nan(X):
    droped_columns = []
    for col in list(X.columns):
            if X[col].isnull().mean() > 0.7:
                droped_columns.append(col)
    X =  X.drop(columns=droped_columns, axis=1)
    return X, droped_columns
```

```python
df, droped_columns = more_70_of_nan(df)
```

```python
df.isnull().mean() * 100
```

```python
#გადავაგდოთ სვეტი "alive" ვინაიდან ის იგივე რაც "survived", ასევე "embarked" არის "embark_town" ,"sex" და "who"
df =  df.drop(['alive','embark_town', 'who'], axis=1)
```

მანქანური დასწავლის ვერც ერთი ალგორითმი ვერ აღიქვამს ტექსტურ ინფორმაციას. შესაბამისად ასეთი ცვლადების არსებობის შემთხვევაში საჭიროა მათი კოდირება რიცხვით ფორმატში. ამ პროცედურითვის გამოვიყენოთ შემდეგი ბიბლიოთეკა:

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
le.fit(df['sex'])
df['sex'] = le.transform(df['sex'])

le.fit(df['embarked'])
df['embarked'] = le.transform(df['embarked'])

le.fit(df['class'])
df['class'] = le.transform(df['class'])

le.fit(df['adult_male'])
df['adult_male'] = le.transform(df['adult_male'])

le.fit(df['alone'])
df['alone'] = le.transform(df['alone'])
```

```python
#გამოვყოტ დამოკიდებული და დამოუკიდებელი ცვლადები
x = df.drop('survived', axis=1)
y = df[['survived']]
```

გავხლიჩოთ მონაცემები საწვრთლენ და სატესტო ნაწილებად:

```python
#გახლეჩვის ბიბლიოთეკა
from sklearn.model_selection import train_test_split, StratifiedKFold

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, shuffle = True, random_state=42, stratify=y)
# stratify=y ნიშნავს საპროგნოზო ცვლადის პროპორციულად გახლეჩვას, 
#ანუ 1 ები და 0 ები სატესტო და საწვრთნელ ნაწილებში იქნება პროპორციული.
```

```python
x_test.isnull().mean() * 100
```

age რიცხვითი ცვლადში გვაქვს 19.24% სიცარიელე, შევავსოთ სიცარიელეები მედიანურ მნიშვნელობებით:

```python
x_train['age'].median()
```

```python
x_train['age'] = x_train['age'].fillna(x_train['age'].median())

#სატესტო მონაცემის სიცარიელე შევავსოთ საწვრთნელი მონაცემის მედიანური მნიშვნელობით:

x_test['age'] = x_test['age'].fillna(x_train['age'].median())
```

embarked კატეგორიული ცვლადში გვაქვს 0.28% სიცარიელე, შევავსოთ სიცარიელეები ახალი კატეგორიით 'unknown':

```python
x_train['embarked'] = x_train['embarked'].fillna('unknown')
```

ჩვენი მონაცემები უკვე მოწესრიგებულია, შეგვიძლია ავაგოთ გადაწყვეტილების ხეების კლასიფიკაციის მოდელი:

```python
#გააწყვეტილების ხეების კლასიფიკასიის ბიბლიოთეკა
from sklearn.tree import DecisionTreeClassifier
#კროსვალიდაციის ბიბლიოთეკები
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
#მოდელი დეფოლტ პარამეტრებით
dtc= DecisionTreeClassifier()
#ფოლდების შექმნა
folds = KFold(n_splits = 5, shuffle = True, random_state = 100)
#მოდელის გაშვება თითოეულ ფოლდზე
scores = cross_val_score(dtc, x_train, y_train, scoring='accuracy', cv=folds)
scores 
```

მივიღეთ თითოეულ ფოლდზე accuracy მეტრიკის მნიშნველობა.

```python
#მოვარგოთ იგივე მოდელი მთლიან საწვრთნელ ნაწილს
dtc.fit(x_train, y_train)
```

ვიპროგნოზოთ სატესტნო მონაცემებზე:

```python
y_pred = dtc.predict(x_test)
```

```python
y_pred_prob = dtc.predict_proba(x_test)
```

## რანდომ ტყეები (Random forest)

გადაწყვეტილების ხეები მარტივი გამაყენებელია, მარტივიად ინტერპრეტირებადი, მაგრამ პრაქტიკაში იც თუ ისე ზუსტი.

ხოლო რადომ ტყეების მოდელებს გააჩნიათ გადაწყვეტილების ხეების სიმარტივე და მოდელის სიზუსტეც იზრდება.

მოდით ავაგოთ რანდომ ტყეები.

ნაბიჯი პირველი: უნდა შევქმნათ "bootstrapped" მონაცემები ძირითადი მონაცემებიდან. ეს პროცესი გამოიყურება შემდეგნაირად:
უნდა მივიღოთ ახალი "bootstrapped" მონაცემები ძირითადი მონაცემებიდან (ძირითადი მონაცემების ზომის) რანდომულად (შემთხვევითად) ვარჩევთ სტრიქონებს, ისე რომ ეს სტრიქონები შეიძლება განმეორდეს:

მნიშვნელოვანი დეტალია, რომ ჩვენ შეგვიძლია ავიღოთ იგივე სტრიქონი ერთზე მეტჯერ.

ნაბიჯი მეროე: ახლად მიღებულ "bootstrapped" მონაცემებზე უნდა შევქმნათ უკვე კარგად ნაცნობი გადაწყვეტილების ხეები, ოღონდ გამოყენებულ უნდა იქნეს შემტხვევიტად არჩეული მხოლოდ ნაწილი სვეტების (ცვლადების) თითოეულ ეტაპზე.

ჩვენ ავაგეთ ხე "ახალ" მონაცემების ქვესიმრავლეზე (რამოდენიმე სვეტზე). 

შემდგომ ეტაპზე თავიდან ვიმერებთ პირველ 2 ნაბიჯს და ვაგებთ მრავალ გადაწყვეტილების ხეს, ანუ რადომულ ტყეს:

როგორ უნდა გამოვიყენოთ ეს ტყე? როდესაც შემოდის ახალი საპროგნოზო მონაცემი, ის გაივლის თითოეულ (მრავალ) აგებულ ხეში და ხეების "ხმათა უმრავლესობით" გამოვლინდება ნაპროგნოზები კლასი.

როგორც გადაწყვეტილების ხეების შემთხვევაში, ანალოგიურად მარტივად შეგვიძია ბიბლიოთეკების გამოყენებით ავაგოთ ეს კომპლექსური მოდელი

```python
#რანდომული ტყეების ბიბლიოთეკა კალსიფიკაციის ამოცანისთვის
from sklearn.ensemble import RandomForestClassifier
```

```python
#მოდელი დეფოლტ პარამეტრებით
rfc= RandomForestClassifier()
#ფოლდების შექმნა
folds = KFold(n_splits = 5, shuffle = True, random_state = 100)
#მოდელის გაშვება თითოეულ ფოლდზე
scores = cross_val_score(rfc, x_train, y_train, scoring='accuracy', cv=folds)
scores
```

```python
#მოვარგოთ იგივე მოდელი მთლიან საწვრთნელ ნაწილს
rfc.fit(x_train, y_train)
#ვიპროგნოზოთ სატესტო მონაცემებზე
y_pred = rfc.predict(x_test)
```

პარამეტრების ოპტიმიზაციისთვის გამოვიყენოთ grid search მეთოდი, ამ მეთოდით ცვენ ვქმნით პარამერების მნიშვნელობებს, საიდანც ყველა შესაძლო კომბინაციაზე იქნება მოდელი გადატვლილი ოპტიმალური კომბინაციის საპოვნელად.
