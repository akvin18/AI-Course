### Task 1: Estimate the Derivative from a Table

- Given the following table of values for f(x), estimate the derivative at x = 2 using the difference quotient:
```
x     f(x)
1.9   3.61
2.0   4.00
2.1   4.41
```
- Use the symmetric difference formula:  
  f'(x) ≈ [f(x + h) - f(x - h)] / (2h)


````python
# Your code here
x = 2.0
h = 0.1
f_x_plus_h = 4.41
f_x_minus_h = 3.61

symmetric_diff = (f_x_plus_h - f_x_minus_h) / (2 * h)
print("Estimated derivative at x = 2:", symmetric_diff)
````

### Task 2: Visualize a Function and Its Tangent

- Plot f(x) = x² using matplotlib.
- Plot the tangent line at x = 2.
- Use the derivative f'(x) = 2x to find the slope at x = 2.


````python
# Your code here
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 100)
y = x**2

# Tangent line at x = 2
x0 = 2
slope = 2 * x0
y0 = x0**2
tangent = slope * (x - x0) + y0

plt.plot(x, y, label='f(x) = x²')
plt.plot(x, tangent, '--', label='Tangent at x=2')
plt.scatter([x0], [y0], color='red')
plt.legend()
plt.title('Function and Tangent Line at x = 2')
plt.grid(True)
plt.show()
````
