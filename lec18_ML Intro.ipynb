# რა არის ML-ი?

ერთ–ერთი მთავარი მიზეზი რის გამოც ვანვითარებთ კომპიუტერულ პროგრამებს არის ის, რომ მოვახდინოთ სხვადასხვა ტიპის პროცესების ავტომატიზაცია. კერძოდ მანქანური დასწავლა წარმოადგენს ხელოვნური ინტელექტის ქვე მიმართულებას რომლის ზოგადი მიზანიც არის კომპიუტერული პრგრამების "ხელით ჩარევის" გარეშე განვითარება. უხეშად შეგვიძლია ვთქვათ, რომ მანქანური დასწავლა არის "ავტომატიზაციის ავტომატიზაცია". სხვა სიტყვებით მანქანური დასწავლა აძლევს კომპიუტერს საშუალებას შექმნას პროგრამები თავად. ML არის მონაცემების პროგრამად გარდაქმნის პროცესი.

მანქანური დასწავლისას ჩვენ საქმე გვაქვს მონაცემებთან რომლებიც შედგებიან მრავალი ჩანაწერისგან, სადაც თითოეული ჩანაწერი წარმოადგენს ჩვენთვის გასაანალიზებელ ობიექტს. მონაცემების შესადგენად საჭიროა ცვლადების შეგროვება და გაზომვა. ეს ცვლადები შეიძლება იყოს კატეგორიული (წინასწარ განსაზღვრული მნიშვნელობები მაგ.: 'ქალი', 'კაცი') ან უწყვეტი (მაგ,: წინასწარ განსაზრვრული მნიშვნელობა" 'ასაკი', 'შემოსავალი'). თითოეული ცვლადი წარმოადგენს ერთ სვეტს ცვლადების ცხრილიდან (მონაცმეებიდან).

ზოგჯერ არსებული მონაცემების ტრანსფორმაცია მნიშვნელოვანი ნაბიჯია მონაცმებების დამუშავების, რამაც შეიძლება მნიშნველოვანი გავლენა იქონიოს მოდელის შედეგზე. შესაბამისად არსებითია ამოცანის სწორი გააზრება და შესაბამისი სფეროს ექსპერტიზა, რათა მოვახერხოთ რელევანტური ცვლადების შექმნა. ამ პროცეს ცვლადების ინჟინერიას უწოდებენ.

მონაცემების ხარისხი უმნიშვნელოვანეს როლს თამაშობს მანქანურ დასწავლაში. მაგალითად ცარიელი ჩანაწერები, რაც ხშირად ახლავს რეალური პრობლემების შესაბამის მონაცემებს, ხშირად შეიძლება იყოს გამოწვევა. მრავალი მიდგომა არსებობს მსგავსი პრობლემის გადასაჭრელად, ხოლო მიდგომის არჩევა კი დამოკიდებულია მონაცემებზე. ტრივიალური მიდგომა ცარიელი მნიშვნელობების შესავსებად არის  სვეტის შესაბამისი საშუალო მნიშვნელობით შევსება.

მანქანური დასწავლა შეიძლება იყოს supervised, unsupervised, semi-supervised, self-supervised, week-supervised  და reinforcement. ჩვენ კურსის ფარგლებში შევეხებით ძირითად პირველ ორ ტიპს.

supervised დასწავლის ამოცანის შემთხვევაში, კომპიუტერს ჩვენ ვაწვდით საწვრთნელ მონაცემებს დაკვირვებებით და შესაბამისი შედეგების მნიშვნელობებით. მიზანი არის რომ ავაგოთ დამოკიდებულება ("მოდელი") დაკვირვებებსა და შედეგებს შორის, რათა შემდგომ შევძლოთ ვიპროგნოზოთ შედეგი ახალ მონაცემებზე, სადაც გვაქვს მხოლოდ დაკვირვებები და არ გაგვაჩნია მათთან დაკავშირებული შედეგები.

დაკვირვებითი (Supervised) სწავლების ამოცანა იყოფა ორი ტიპის ამოცანად:

    1. კლასიფიკაცია – სადაც შედეგების მნიშვნელობა არის კატეგორიული. 

ამ შემთხვევაში ჩვენ გვაქვს ორი კლასი და ამოცანა არის ისე ვასწავლოთ "მოდელს", რომ მას შემდგომ შეეძლოს  ამ ორი კლასის განსხვავება გარკვეული სიზუსტით.

    2. რეგრესია –  სადაც შედეგების მნიშვნელობა არის რიცხვითი.

ეხლა ჩვენ გვაქვს მოყვანილი უმარტივესი ერთი ცვლადის (x) წრფივი რეგრესიის მოდელი საპროგნოზო ცვლადი y-ით.
ამოცანა კი არის შემდეგი რომ გავავლოთ ისეთი წრფე (მოდელი) რომ მომავალში ახლად შემოსული მონაცემები საშუალოდ მინიმალურად იყვნენ დაშორებული ამ წრფიდან.

ფაქტიურად ეს ორი იდეა არის მანქანრუი დასწავლის (და თამამად შეგვიძლია ვთქვათ ხელოვნური ინტელექტის ფუნდამენტი), თუ კი ეს კონცეფცია გვესმის კარგად შეგვიძლია ჩავთვალოთ, რომ ჩვენ მანქანური დასწავლის და ხელოვნური ინტელექტის უზარმაზარი სფეროს კარები შეღებული გვაქვს.

შემდეგი კატეგორიის ამოცანა არის დაკვირვების გარეშე (unsupervised)  მაქნანური დასწავლა, რაც გულისხმობს, რომ ჩვენ საქმე გვაქვს მონაცემებთან რომელთაც label–ები არ გააჩნიათ და ჩვენ გვსურს მონაცემებში ბუნებრივად წარმოშობილი კანონზომიერების პოვნა. ყველაზე გავრცელებული unsupervised მანქანური დასწავლის ამოცანა არის კლასტერების შექმნა:

კლასტერირების მეთოდის გამოყენების მიზანია, რომ მოვახდინოთ მონაცემებში რელევანტური ქვეჯგუფების გამოყოფა, ქვეჯგუფების შესახებ წინასწარ განსაზღვრული ჰიპოტეზის გარეშე.

მანქანური დასწავლის ალგორითმის მიზანია ისეთი შესაბამისობის (მოდელის) სწავლა რომელიც აღწერს დაკვირვებად საწვრთნელ მონაცემებს და შეიძლება მისი განზოგადოება ახალ დამოუკიდებელ მონაცემებზე.

ქვემოთ შევეხებით დაკვირვებითი დასწავლის მოდელის ზედმეტად მორგებულობის და უვარგისობის საკითხს. შემდგომ ავხსნით მოდელის წარმადობის შეფასებას კლასიფიკაციის და რეგრესიის შემთხვევაში. სხვადასხვა წარმადობის საზომების (მეტრიკების) სწორად გაგება გვაძლევს საშუალებას უკეთ შევაფასოთ თავად მოდელი.

მანქანური დასწავლის ალორითმის მიზანია ისწავლოს კანონზომიერება დაკვირვებად საწვრთნელ მონაცემებზე და შემდგომ შევძლოთ მისი განზოგადება ახალ დამოუკიდებელ ცვლადებზე ის რომ ავირიდოთ თავიდან მოდელის ზემდეტად მორგებულობა და მოდელის უვარგისობა.

მოდელმა უნდა დააკმაყოფილოს ორი პირობა:
 1. საწვრთნელ მონაცემებზე მოდელმა უნდა იპროგნოზოს მაღალი სიზუსტით დაკვირვებადი მნიშვნელობები, წინააღმდეგ   შემთხვევაში ვიტყვით, რომ მოდელი უვარგისია და გაჩნია მაღალი მიკერძოებულობა.
 2. მოდელის განზოგადოება უნდა შეგვეძლოს საწვრთნელი მონაცემების მიღმა. ზედმეტად მორგებული მოდელი კარგად პროგნოზირებს საწვრთნელ მონაცემებზე, მაგრამ ცუდად პროგნოზირებს დამოუკიდებელ სატესტო მონაცემებზე, ხშირად ეს გამოწვეულია მოდელის კომპლექსურობიდან. ასეთ დროს გვაქვს მაღალი ვარიანსი. 

overfitting – მოდელი კარგად პრპგნოზირებს საწვრთნელ მონაცემებზე მაგრამ სატესტო, "უნახავ" მონაცემებზე აქვს შედარებით ცუდი შედეგი. 

ამ სურათზე ნაჩვენებია რეგრესიის შემთხვევაში ზედმეტად მორგებული (overfitting) და უვარგისი (underfitting) მოდელების პრობლემა. Underfitting ხდება მაშინ როდესაც მოდელი ძალიან მარტივია ან ცვლადები რისგანაც შედგება მონაცემებ არ არიან სკმარისად ინფორმატიულები. Overfitting ხდება მაშინ როდესაც მოდელი ძალინა კომპლექსურია.

მიკერძოებულობა (bias) გვაჩვენებს მოდელის საშუალო ცდომილებას სხვადასხვა საწვრთნელი მონაცემისთვის: ეს არის შეუსაბამობა საშუალო ნაპროგნოზევ მნიშვნელობასა და რეალურ საშუალოს შორის რომლის პროგნოზირებასაც ვცდილბთ.

ვარიანსი გვაჩვენებს მოდელის სენსიტიურობას საწვრთნელ მონაცემებთან: მოცემული მნიშვნელობისთვის ის წარმოადგენს ნაპროგნოზები მნიშვნელობის გავრცელებას მათი საშუალოს გარშემო. 

ზედმეტად მორგების პრობლემა აჩვენებს რომ მოდელის შესაფასებლად არ არის კარგი ინდიკატორი მისი წარმადობის შეფასება საწვრთნელ მონაცემებზე.

მსგავსი საფრთხეების თავიდან ასარიდებლად გამოყენება მონაცემის 3 ნაწილად გაყოფა
1. საწვრთნელი ნაწილი– სადაც ჩვენი მოდელი სწავლობს კანონზომიერებებს
2. ვალიდაციის ნაწილი – სადაც ვამოწმებთ თუ რამდენად კარგად ისწავლა ჩვენმა მოდელმა 
3. სატესტო ნაწილი– სადაც ვცდით თუ როგორი წარმადობა ექნება მოდელს მომავალში სრულიად უნახავ მონაცემებზე

მაგ. თუ მონაცემების წარმადობა საწვრთნელ და ვალიდაციის ნაწილებზე ერთნაირად კარგია (ანუ არ გვიდგას overfitting-ის პრობლემა) ჩვენ გვაქვს უფლება სატესტო ნაწილზე შევამოწმოთ მოდელის წარმადობა. მიღებული შედეგის საფუძველზე შევძლებთ დასკვნა გავაკეთოთ მოდელის ავკარგიანობაზე

თუმცა ვალიდაციის უფრო აპრობირებული მეთოდია კროსვალიდაცია, რაც მდგომარეობს შემდეგში:

როგორც ხედავთ ვალიდაციას ვახდენთ ამ შემთხვევაში 10 ჯერ (საწვრთნელი მონაცემი იყოფა 10 ტოლ ნაწილად სადაც პირველი 1/10 არის ვალიდაციის ნაწილი ხოლო დანარჩენი 9/10 არის საწვრთნელი, შემდგომ მეორე 1/10 არის ვალიდაციის ნაწილი და ა.შ.  მიზეზი თუ რატომ ვაკეთებთ ამ პროცედურას არის ის რომ ვინაიდან ვალიდაციის ნაწილის მოჩრა არის შემტხვევით (random) პროცესი შეიძლება სხვადასხვა იტერაციაზე განსხვავებული შედეგი მივიღოთ, ასეთი დახლეცვით კი  და თითოეულ ნაჭერზე (fold) ზე მიღებული შედეგების საშუალოს აღება გაცილებით ზუსტ წარმოდგენას შეგვიქმნის მთლიანად მონაცემებზე მომუშავე მოდელის შესახებ.

## მოდელის წარმადობის მეტრიკები

რეგრესიის ამოცანების წარმადობის შესაფასებლად ვიყენებთ პროგრნოზირებულ და რეალურ მნიშნვლნელობებს შორის საშუალო მანძილებს. 

მაგალითად ეს მანძილები შეიძლება დაითვალოს ევკლიდური მანძილით. 


აქვე გავიხსენოთ ორ წერტილს შორის ევკლიდური მანძილის განმარტება:

სადაც

მსგავსი მანძილით წარმოშობილ მეტრიკას ვუწოდებთ RMSE-ს (root mean square error).


ანალოგიურად შეგვიძია განვმარტოთ მეტრიკა აბსოლუტური მნიშვნელობებით კვადრატების გარეშე, რომელსაც ეწოდესბა MAE (mean absolute error). აღნიშნული მეტრიკა, პროგრნოზირებულ და რეალურ მნიშნვლნელობებს შორის საშუალო აბსოლუტურ გადახრას გვიჩვენებს.

ბინარული კლასიფიკაციის ამოცანების შემთხვევაში წარმადობის მეტრიკები გამოგვყავს “confusion matrix”–დან.

მოდით დავიწყოთ “confusion matrix”–ის ახსნა ბინარული კლასიფიკაციის მაგალითზე (ცხადია ეს ყველაფერი შეიძლება განზოგადდეს მრავალი კლასის კლასიფიკაციის შემთხვევებშიც). 

რა შეგვიძლია გავიგოთ ამ მატრიციდან?

1.  გვაქვს ორი შესაძლო საპროგნოზო კლასი: "კი" და "არა". მაგალითად თუ ვიპროგნოზებდით კლიენტის გადეფოლტებას ბანკში, "კი" კლასი იქნებოდა გადეფოლტება, ხოლო "არა" არა გადეფოლტებულთა კლასი.
 
2. კლასიფიკაციის მოდელმა სულ ჯამში 165 პროგნოზი მოგვცა (ანუ 165 კლიენტზე შემოწმდა გადეფოლტდებიან თუ არა ისინი).

3. ამ 165 შემთხვევიდან, 110 "კი" ვიპროგნოზეთ და 55 "არა".

4. სინამდვილეში 105 კლიენტი არის გადეფოლტებული, ხოლო 60 არ არის.

შემოვიტანოთ საბაზისო ტერმინები:

* true positives (TP) / ჭეშმარიტი დადებითები : ეს არის ის შემთხვევები სადაც ჩვენ ვიპროგნოზეთ "კი" და სინამდვილეშიც არის "კი".

* true negatives (TN) / ჭეშმარიტი უარყოფითები : ეს არის ის შემთხვევები სადაც ჩვენ ვიპროგნოზეთ "არა" და სინამდვილეშიც არის "არა".

* false positives (FP) / მცდარი დადებითები : ეს არის ის შემთხვევები სადაც ჩვენ ვიპროგნოზეთ "კი" და სინამდვილეში არის "არა".

* false negatives (FN) / მცდარი უარყოფითები : ეს არის ის შემთხვევები სადაც ჩვენ ვიპროგნოზეთ "არა" და სინამდვილეში არის "კი".





შესაბამისად ჩვენი "confusion matrix" გამოიყურება შემდეგნაირად:

მეტრიკები რომლებიც ბინარული კლასიფიკაციისთვის ითვლება confusion matrix–დან:

1. Accuracy: მთლიანობაში რამდენად ზუსტად იცნობს მოდელი?


    (TP+TN)/სულ = (100+50)/165 = 0.91
    
    
2. Precision: როდესაც ვპროგნოზირებთ "კი"–ს, რამდენად სწორად ვპროგნოზირებთ მას?


     TP/ნაპროგნოზები "კი" = 100/110 = 0.91    
     

3. Recall: არსებული "კი"–ებიდან რამდენად სწორად ვპორგნოზირებთ "კი"–ს?  (გივე რაც True Positive Rate, Sensitivity)

    
     TP/არსებული "კი" = 100/105 = 0.95 
   

4. Specificity (True Negative Rate): ებული "არა"–ებიდან რამდენად სწორად ვპროგნოზირებთ "არა"–ებს? 


     TN/არსებული "არა" = 50/60 = 0.83


ROC Curve: ეს გრაფიკი ხშირად გამოიყენება კლასიფიკაციის მოდელის წარმადობის შესაფასებლად ყველა შესაძლო ზღვარებზე (thresholds) (ვინაიდან კლასიფიკაციის მოდელი აბრუებს კლასის ალბათობას, სწორედ ეს "ზღვარი" განსაზღვრავს თუ რომელ კლას მივაკუთვნებთ პრგნოზს).


ROC წირი მიიღება True Positive Rate და False Positive Rate შესაბამისი გრაფიკის აგებით კონკრეტული thresholds–სთვის.


ზოგადად thresholds-ის არჩევა "ბიზნეს გადაწყვეტილებაა": მოვახდინოთ False Positive Rate–ის მინიმიზაცია თუ მოვახდინოთ True Positive Rate–ის მაქსიმიზაცია.
