- A) The function is at a minimum or maximum  
- B) The function is increasing

- A) Maximum decrease  
- B) Constant output

**7. The second derivative of a function tells us about:**

- A) The function is non-differentiable

**10. A function is differentiable if:**

**11. A matrix is best described as:**  
- A) A list of random values

**19. The rank of a matrix represents:**  
- A) The total number of elements

**20. In general, the gradient of a multivariable function tells us:**
