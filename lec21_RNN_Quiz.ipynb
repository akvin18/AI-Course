# Quiz: Recurrent Neural Networks (RNNs)

### Multiple Choice Questions

1. What is the key feature that distinguishes RNNs from traditional feedforward neural networks?  
a) Use of dropout  
b) Use of attention mechanisms  
c) Ability to process sequential data with memory  
d) Activation function choice

2. Which of the following problems is most appropriate for an RNN?  
a) Image classification  
b) Stock price prediction over time  
c) Spam detection  
d) Object detection in images

3. What is the vanishing gradient problem in RNNs?  
a) A gradient that becomes zero when learning finishes  
b) Gradients become too small during backpropagation through time  
c) Gradients overflow and become too large  
d) Gradient noise from dropout layers

4. What is a common technique to address the vanishing gradient issue in RNNs?  
a) Add more layers  
b) Use smaller datasets  
c) Use LSTM or GRU units  
d) Increase the learning rate

5. How does an LSTM differ from a vanilla RNN?  
a) It uses convolution instead of matrix multiplication  
b) It introduces gates to control memory  
c) It doesn’t require backpropagation  
d) It is only used for image data


### Analytical Questions

1. Why might a standard feedforward neural network fail on sequence prediction tasks, and how do RNNs address this limitation?

2. Explain how hidden states are passed from one time step to the next in an RNN. What kind of dependencies does this enable?

3. Describe a scenario where using an LSTM would be better than a simple RNN.

4. What is the role of the forget gate in an LSTM? What happens if the forget gate always outputs 1? What if it outputs 0?

5. RNNs are known to struggle with long-term dependencies. What are some practical strategies or architectural changes to overcome this?
