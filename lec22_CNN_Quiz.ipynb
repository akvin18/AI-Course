# Quiz: Convolutional Neural Networks (CNNs)

### Multiple Choice Questions

1. What is the main purpose of a convolutional layer in a CNN?  
a) To flatten the input data  
b) To apply random noise to the image  
c) To extract spatial features using filters  
d) To reduce dimensionality using PCA

2. What does a pooling layer do in a CNN?  
a) Increases resolution  
b) Normalizes pixel values  
c) Reduces spatial dimensions and computations  
d) Applies non-linearity

3. Which of the following statements is true about CNN filters?  
a) Filters are hand-coded based on visual patterns  
b) Filters are learned during training  
c) Filters are random noise  
d) Filters are used only in the final layer

4. What is a drawback of using only fully connected layers on image data?  
a) They work faster than convolutional layers  
b) They generalize better than CNNs  
c) They ignore spatial relationships in pixels  
d) They work better on grayscale images

5. Which activation function is most commonly used in modern CNNs?  
a) Sigmoid  
b) Tanh  
c) ReLU  
d) Softmax
